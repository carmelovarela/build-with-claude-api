```python
# Create an API client

from anthropic import Anthropic

client = Anthropic()
model = "claude-sonnet-4-0"
```

```python
def add_user_message(messages, text):
    user_message = {"role": "user", "content": text}
    messages.append(user_message)

def add_assistant_message(messages, text):
    assistant_message = {"role": "assistant", "content": text}
    messages.append(assistant_message)

def chat(messages):
    message = client.messages.create(
        model=model,
        max_tokens=1000,
        messages=messages
    )
    return message.content[0].text
```

```python
# Make a starting list of messsages
messages = []

# Add in the initial user question of "Define quantitative research in one sentence"
add_user_message(messages, "Define quantitative research in one sentence")

# Pass the list of messsages into 'chat' to get an answer
answer = chat(messages)

# Take the answer and add it as an assistant message into our list
add_assistant_message(messages, answer)

messages
# Add in the user's follow-up question
add_user_message(messages, "List the key steps in a quant research workflow")

# Call that again with the list of messages to get a final answer 
answer = chat(messages)
answer
```
